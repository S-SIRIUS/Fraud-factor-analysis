# 용의자 데이터 카이스퀘어 검정
## 유의확률이 매우낮음 -> 귀무가설을 기각한다. -> 두 변수는 연관성이 있다.
## 지역과 나이는 연관성이 있다.

```python
df_suspect_origin
```

```python
suspect_crosstab=pd.crosstab(df_suspect_origin['광역시도명'], df_suspect_origin['생년구간'], margins=False)
```

```python
suspect_crosstab
```

```python
#교차분석 결과값들 일괄계산
#correction=True를 적용하면 Yates' correction이 적용 되어 검정통계량이 보수적으로 더 낮게 나옴
result=chi2_contingency(observed=suspect_crosstab, correction=False)
print("1. 카이제곱 통계량:", result[0])
print("2. p-value:", result[1])
print("3. df:", result[2]) #(행의개수-1)*(열의개수-1)
print("4. 기대값 행렬:")
```

# 생년구간도 변경

```python
def change_age(row):
    if row=='1900-1910':
        return '1900-1920'
    elif row=='1910-1920':
        return '1900-1920'
    elif row == '1930-1940':
        return '1930-1950'
    elif row == '1940-1950':
        return '1930-1950'
    elif row == '1950-1960':
        return '1950-1970'
    elif row == '1960-1970':
        return '1950-1970'
    elif row == '2000-2010':
        return '2000-2020'
    elif row == '2010-2020':
        return '2000-2020'
    else:
        return row
    
```

```python
df_suspect_origin['new생년구간']=df_suspect_origin['생년구간'].apply(change_age)
```

```python
suspect_crosstab=pd.crosstab(df_suspect_origin['도시'], df_suspect_origin['new생년구간'], margins=False)
```

```python
suspect_crosstab
```

```python
suspect_crosstab
```
